```python
import pandas as pd
import math
import numpy as np
data = pd.read_csv('1KMinData_TXF.csv')
```

```python
DateTime = pd.DataFrame.as_matrix(data['dDateTime'])
Close = pd.DataFrame.as_matrix(data['dClose'])
Open = pd.DataFrame.as_matrix(data['dOpen'])
High = pd.DataFrame.as_matrix(data['dHighest'])
Low = pd.DataFrame.as_matrix(data['dLowest'])
Volume = pd.DataFrame.as_matrix(data['dVolume'])
```

```python
ma = 15
a = 2
b = 3
c = 5
d = 20
```

```python
# MA、MV計算
size = len(DateTime)
MAR = np.zeros((size,1),dtype = np.float)
MV = np.zeros((size,1),dtype = np.float)
for i in range (ma,len(DateTime)):
    if (int(DateTime[i]/10000) == int(DateTime[i-ma]/10000)):
        MAR[i] = np.mean(abs(Close[i-ma:i]-Open[i-ma:i]))
        MV[i] = np.mean(Volume[i-ma:i])

index = []
target = []
#判斷是否符合條件
for i in range(size-d):
    if (MAR[i] != 0): #確定日內
        if (Volume[i]> MV[i]*a and (abs(Close[i]-Open[i]))> MAR[i]*b): #符合帶大量 大漲/大跌
            index.append(i)
            #將符合條件者分類
            if (Close[i]-Open[i]) > 0: #漲
                if (Close[i+d]-Open[i+c+1]) > 0 and Close[i+d]>High[i]: #同向 最後高於高點
                    target.append(0)
                elif (Close[i+d]-Open[i+c+1]) < 0 and Close[i+d]<Low[i]: #反向 最後低於低點
                    target.append(1)
                else:
                    target.append(2)
            else: #跌
                if (Close[i+d]-Open[i+c+1]) < 0 and Close[i+d]<Low[i]: #同向 最後低於低點
                    target.append(0)
                elif (Close[i+d]-Open[i+c+1]) > 0 and Close[i+d]>High[i]: #反向 最後高於高點
                    target.append(1)
                else:
                    target.append(2)
    
```

```python
#乖離率作圖
def biasImage(Data):
    Image = np.zeros((20,20),dtype = np.float)
    for i in range(20):
        for j in range(2,22):
            avg = np.mean(Data[i:i+j])
            Image[i][j-2] = (Data[i+j-1]-avg)/avg*100
    
    Max = np.max(Image[:,:])
    Min = np.min(Image[:,:])
    Image = (Image-Min)/(Max-Min)
    return Image
```

```python
a = biasImage(Close[20:20+40])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(a,cmap='gray')
plt.show()
```

```python
#訓練集 && 測試集
XX = np.zeros((len(index),20,20,2),dtype = np.float)
YY = np.zeros((len(index),3),dtype = np.float)

for i in range (len(index)):
    I = index[i]
    XX[i,:,:,0] = biasImage(Close[I-39:I+1])
    XX[i,:,:,1] = biasImage(Volume[I-39:I+1])
    if target[i]==0:
        YY[i,0] = 1
    elif target[i]==1:
        YY[i,1] = 1
    else:
        YY[i,2] =1
```

```python
import pickle
pickle.dump( XX, open( "bias_P&V_15\2\3\5\20_X", "wb" ) )
pickle.dump( YY, open( "bias_P&V_15\2\3\5\20_Y", "wb" ) )
```
